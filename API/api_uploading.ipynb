```python
#  import dependencies
from sqlalchemy import create_engine
import pandas as pd
```

```python
# bring CSVs to this notebook, convert back to df`
teams_df = pd.read_csv(r"C:\Users\hmwhi\OneDrive\Desktop\GitHub\Project2_Team2\API\teams_current_data.csv")
teams_df = teams_df.set_index('team_id')
teams_df.head()
```

```python
# bring CSVs to this notebook, convert back to df`
players_df = pd.read_csv(r"C:\Users\hmwhi\OneDrive\Desktop\GitHub\Project2_Team2\API\players_current_data.csv")
players_df = players_df.set_index('player_id')
players_df.head()
```

```python
# bring CSVs to this notebook, convert back to df`
games_df = pd.read_csv(r"C:\Users\hmwhi\OneDrive\Desktop\GitHub\Project2_Team2\API\games_current_data.csv")
games_df = games_df.drop(columns='Unnamed: 0').set_index('game_id')
games_df.head()
```

```python
engine = create_engine('postgresql://postgres:Timeex00@localhost:5432/mybasketball')
teams_df.to_sql('teams_current_data', engine, if_exists='append')
players_df.to_sql('players_current_data', engine, if_exists='append')
games_df.to_sql('games_current_data', engine, if_exists='append')
```

```python
# double checking that data has been transferred to the SQL database
teams_data = engine.execute('SELECT * FROM teams_current_data LIMIT 3')
for record in teams_data:
    print(record)
```

```python
players_data = engine.execute('SELECT * FROM players_current_data LIMIT 3')
for record in players_data:
    print(record)
```

```python
games_data = engine.execute('SELECT * FROM games_current_data LIMIT 3')
for record in games_data:
    print(record)
```
